```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

$$y'' + y = 0\qquad y(0) = 0; \ y'(0) = 1$$
$$y = \sin x$$
Перепишем данный дифф. уравнение в виде системы, используя замену $y' = t$.
$$y' = t$$
$$t' = -y$$


```python
def solver(x: float, r: tuple) -> tuple:
    y, t = r
    return t, -y

sol = solve_ivp(fun=solver, t_span=(-np.pi, np.pi), y0=(0, 1), t_eval=np.linspace(-np.pi, np.pi, 1000))
y, t = sol.y
X = sol.t
```

```python
plt.plot(X, y)
plt.show()
```
